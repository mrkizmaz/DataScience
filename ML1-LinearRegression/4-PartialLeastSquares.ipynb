### Verisetinin İncelenmesi ve Manipüle Edilmesi

* Kullanılan Kütüphaneler

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression, PLSSVD

import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

* Verisetinin İncelenmesi

Verisetini [buradan](https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv) indirebilirsiniz.

```python
hit = pd.read_csv("https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv")
df = hit.copy()
df.head()
```

```python
df.info() # veriseti hakkında bilgi verir
```

```python
df.isnull().values.any() # verisetinde bos degerler var mı?
```

```python
df = df.dropna(axis = 0) # verisetindeki bos degerlerin oldugu satırları siler
```

```python
df.info()
```

```python
df.nunique() # sütun bazında eşsiz degerlerin sayısı
```

```python
df.describe().T # verisetinin istatistiksel dagılımı
```

```python
df.corr() # aralarındaki korelasyon degerleri
```

```python
# verisetindeki kategorik degiskenler sürekli degiskenlere cevrilmeli,
dummy = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
dummy.head()
```

```python
# kategorik degiskenler ana verisetinden silinmeli,
df1 = df.drop(['League', 'Division', 'NewLeague'], axis = 1)
df1.head()
```

```python
# dummy ve df1 birlestirerek güncel veriseti olusturulur,
df = pd.concat([df1, dummy[['League_N', 'Division_W', 'NewLeague_N']]], axis = 1)
df.head()
```

### Model Kurulumu

```python
# %80 egitim, %20 test seti,
X = df.drop(['Salary'], axis =  1) # Salary dısındaki bagımsız degiskenler
y = df['Salary'] # bagımlı degisken

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.20,
                                                   random_state = 42)

pls = PLSRegression() # bilesen sayısı girilmedi! (n_components = 2, 4, 6, ... girilebilir)
pls_model = pls.fit(X_train, y_train)
pls_model
```

```python
pls_model.coef_[0:5] # olusturulan modelin katsayıları
```

```python
rscore = pls_model.score(X_train, y_train) # modelin basarı oranı
```

```python
rscore # [önemli]
```

```python
# bilesen sayısı girilirse,
pls2 = PLSRegression(n_components = 4)
pls_model2 = pls2.fit(X_train, y_train)
pls_model2
```

```python
pls_model2.coef_[0:5]
```

```python
rscore2 = pls_model2.score(X_train, y_train) # modelin basarı degeri
```

```python
rscore2 # [önemli]
```

### Tahminleme

```python
# bilesen sayısı girilmeden elde edilen hata degeri
ypred = pls_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, ypred))
rmse # hata degeri [önemli]
```

```python
# bilesen sayısı 4 oldugunda elde edilen hata degeri
ypred2 = pls_model2.predict(X_test)
rmse2 = np.sqrt(mean_squared_error(y_test, ypred2))
rmse2 # hata degeri [önemli]
```

```python
print("Bilesen sayısı girilmeden elde edilen degerler: \nHata degeri = {0} \nModelin basarı yüzdesi = {1}".format(rmse, rscore * 100))
print("Bilesen sayısı 4 iken elde edilen degerler: \nHata degeri = {0} \nModelin basarı yüzdesi = {1}".format(rmse2, rscore2 * 100))
```

* Görüldügü üzere bu sekilde optimum hata degeri elde edilemez, bilesen sayısı degistikce hata degerleri de degisecektir. (n_compenents = 2, 4, 5, 10, 20, ...)

### Model Tuning / Model Doğrulama

```python
# 10 katlı cross validation ile model dogrulama,

cv10 = model_selection.KFold(n_splits = 10, shuffle = True, random_state = 1)

RMSE = [] # hesaplanan hata degerleri listesi

for i in np.arange(1, X_train.shape[1] + 1):
    pls_reg = PLSRegression(n_components = i)
    model = pls_reg.fit(X_train, y_train)
    score = np.sqrt(-1 * model_selection.cross_val_score(pls_reg,
                                                          X_train,
                                                          y_train,
                                                          cv = cv10,
                                                          scoring = 'neg_mean_squared_error').mean())
    RMSE.append(score)
    
# Degerlerin Görsellestirilmesi,
plt.plot(np.arange(1, X_train.shape[1] + 1), RMSE, '-v', c = 'blue')
plt.title('PLC modeli ile RMSE degerleri')
plt.xlabel('Bilesen Sayısı')
plt.ylabel('RMSE Degerleri');
```

```python
min(RMSE) # minimum hata degeri (optimum deger) [önemli]
```

```python
# optimum hata degerindeki bilesen sayısı nedir?
pls_params = {'n_components': np.arange(1, X_train.shape[1] + 1)}
pls = PLSRegression()
plsmodel = GridSearchCV(pls, pls_params, cv = 10)
modelcv = plsmodel.fit(X_train, y_train)
```

```python
modelcv
```

```python
modelcv.best_params_ # optiumum hata degerinde bilesen sayısı
```

```python
bilesen = modelcv.best_params_["n_components"]
```

```python
bilesen
```

```python
print(f"Cross Validation uygulandıktan sonra elde edilen optimum hata degeri: {min(RMSE)} ve Bilesen sayısı: {bilesen}")
```

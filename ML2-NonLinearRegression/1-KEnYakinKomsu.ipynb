# 1- K En Yakın Komşu (K-Nearest Neighbors)

## TEORİ

**K-NN algoritması, yeni durum/veriler ile mevcut durumlar arasındaki benzerliği varsayar ve yeni durumu mevcut kategorilere en çok benzeyen kategoriye yerleştirir.**

* Denetimli Öğrenme tekniğine dayalı en basit makine öğrenimi algoritmalarından biridir.
* Parametrik olmayan bir algoritmadır, yani temel veriler üzerinde herhangi bir varsayımda bulunmaz.
* Anlaşılması kolaydır ve genelde büyük verisetlerinde kullanılmaz.
* Sınıflandırmanın yanı sıra Regresyon için de kullanılabilir ancak daha çok Sınıflandırma problemleri için kullanılır.

* KNN Basamakları,
    1. Komşu sayısı (k) belirlenir,
    2. Bilinmeyen nokta ile diğer tüm noktalalar arasındaki uzaklıklar hesapalnır,
    3. Uzaklıklar sıralanır ve belirlenen k sayısına en yakın olan gözlem seçilir,
    4. Sınıfılandırma ise en sık sınıf, regresyon ise ortalama değeri tahmin değeri olarak verilir.

## UYGULAMA

### Verisetinin İncelenmesi ve Manipüle Edilmesi

* Kullanılan Kütüphanler

```python
import pandas as pd
import numpy as np
import seaborn as sns
import mglearn
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

* Verisetinin İncelenmesi

Verisetini [buradan](https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv) indirebilirsiniz.

```python
hit = pd.read_csv("https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv")
df = hit.copy()
df.head()
```

```python
df.info() # veriseti hakkında bilgi verir
```

```python
df.isnull().values.any() # verisetinde bos degerler var mı?
```

```python
df = df.dropna(axis = 0) # verisetindeki bos degerlerin oldugu satırları siler
```

```python
df.info()
```

```python
df.nunique() # sütun bazında eşsiz degerlerin sayısı
```

```python
df.describe().T # verisetinin istatistiksel dagılımı
```

```python
df.corr() # aralarındaki korelasyon degerleri
```

```python
# verisetindeki kategorik degiskenler sürekli degiskenlere cevrilmeli,
dummy = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
dummy.head()
```

```python
# kategorik degiskenler ana verisetinden silinmeli,
df1 = df.drop(['League', 'Division', 'NewLeague'], axis = 1)
df1.head()
```

```python
# dummy ve df1 birlestirerek güncel veriseti olusturulur,
df = pd.concat([df1, dummy[['League_N', 'Division_W', 'NewLeague_N']]], axis = 1)
df.head()
```

```python
# örnegin, komsu sayısı 5 olan knn gösterimi,
mglearn.plots.plot_knn_regression(n_neighbors = 4)
```

### Model Kurulumu

```python
X = df.drop(['Salary'], axis =  1) # Salary dısındaki bagımsız degiskenler
y = df['Salary'] # bagımlı degisken

# %80 egitim, %20 test seti,
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.20,
                                                    random_state = 42)
# model nesnesinin olusturulması ve fit edilmesi
knn = KNeighborsRegressor()
knn_model = knn.fit(X_train, y_train)
knn_model
```

```python
knn_model.n_neighbors # belirlenen komsu sayısı verir
```

```python
knn_model.effective_metric_ # uzaklıgın hesaplanmasında kullanılan yöntemi verir
```

```python
rscore = knn_model.score(X_train, y_train) # modelin basatı skoru
```

```python
rscore # [önemli]
```

### Tahminleme

```python
# k = 5 icin hata eegerinin bulunması,
y_pred = knn_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
rmse # [önemli]
```

```python
# farklı k degerlerin hata degerlerinin bulunması,
RMSE = []

for k in range(10):
    k = k + 1
    model = KNeighborsRegressor(n_neighbors = k).fit(X_train, y_train)
    y_pred2 = model.predict(X_test)
    rmse2 = np.sqrt(mean_squared_error(y_test, y_pred2))
    RMSE.append(rmse2)
    print("k = ", k, "icin RMSE degeri: ", rmse2)
```

```python
min(RMSE) # [önemli]
```

```python
print("Model basarı yüzdesi: ", rscore * 100)
print(f"k = 5 icin elde edilen hata degeri: {rmse}")
print(f"Farklı k degerlerinde elde edilen minimum hata degeri: {min(RMSE)}")
```

* Optimum hata degerine ulasılması icin model tuning uygulanmalı

### Model Tuning / Model Doğrulama

```python
# optimum hata degerinin bulunması,
knn_params = {'n_neighbors': np.arange(1, 50, 1)}
```

```python
knn = KNeighborsRegressor().fit(X_train, y_train)
knn_cv_model = GridSearchCV(knn, knn_params, cv = 10)
knn_cv_model = knn_cv_model.fit(X_train, y_train)
```

```python
knn_cv_model.best_params_ # en iyi parametreler
```

en iyi komsu sayısını 3 olarak buldu

```python
# final modelinin olusturulması ve optimum hata degerinin bulunması,
knn_tuned = KNeighborsRegressor(n_neighbors = knn_cv_model.best_params_['n_neighbors'])
knn_tuned = knn_tuned.fit(X_train, y_train)
ypred = knn_tuned.predict(X_test)
rmse_tuned = np.sqrt(mean_squared_error(y_test, ypred))
```

```python
rmse_tuned # [önemli]
```

```python
print("En iyi k degeri icin optimum hata degeri: ", rmse_tuned)
```

* Kullanılan Kütüphaneler

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from skompiler import skompile
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

* Verisetinin İncelenmesi

Verisetini [buradan](https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv) indirebilirsiniz.

```python
hit = pd.read_csv("https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv")
df = hit.copy()
df.head()
```

```python
df.info() # veriseti hakkında bilgi verir
```

```python
df.isnull().values.any() # verisetinde bos degerler var mı?
```

```python
df = df.dropna(axis = 0) # verisetindeki bos degerlerin oldugu satırları siler
```

```python
df.info()
```

```python
df.nunique() # sütun bazında eşsiz degerlerin sayısı
```

```python
df.describe().T # verisetinin istatistiksel dagılımı
```

```python
df.corr() # aralarındaki korelasyon degerleri
```

```python
# verisetindeki kategorik degiskenler sürekli degiskenlere cevrilmeli,
dummy = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
dummy.head()
```

```python
# kategorik degiskenler ana verisetinden silinmeli,
df1 = df.drop(['League', 'Division', 'NewLeague'], axis = 1)
df1.head()
```

```python
# dummy ve df1 birlestirerek güncel veriseti olusturulur,
df = pd.concat([df1, dummy[['League_N', 'Division_W', 'NewLeague_N']]], axis = 1)
df.head()
```

### Model Kurulumu

```python
X = df.drop(['Salary'], axis =  1) # Salary dısındaki bagımsız degiskenler
y = df['Salary'] # bagımlı degisken



# %80 egitim, %20 test seti,
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.20,
                                                    random_state = 42)
```

```python
X_train = pd.DataFrame(X_train['Hits'])
X_test = pd.DataFrame(X_test['Hits'])
```

```python
# modelin olusturulması ve fit edilmesi,
cart_model = DecisionTreeRegressor().fit(X_train, y_train)
cart_model
```

```python
rscore = cart_model.score(X_train, y_train) # model basarı degeri
```

```python
rscore # [önemli]
```

```python
# agac yapısının görsellestirilmesi
X_grid = np.arange(min(np.array(X_train)), max(np.array(X_train)), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X_train, y_train, color = 'r')
plt.plot(X_grid, cart_model.predict(X_grid), color = 'b');
```

### Tahminleme

```python
# hata degerinin bulunması,
y_pred = cart_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
rmse # [önemli]
```

```python
print(f"Agac sayısı girilmeden elde edilen modelin hata degeri: {rmse} ve basarı yüzdesi: {rscore * 100}")
```

### Model Tuning / Model Doğrulama

```python
# optimum degerlerin bulunması,
cart_params = {'min_samples_split': range(2, 100),
              'max_leaf_nodes': range(2, 10)}
```

```python
# modelin olusturulması ve fit edilmesi
cart_cv_model = GridSearchCV(cart_model, cart_params, cv = 10, n_jobs = -1)
cart_cv_model.fit(X_train, y_train) # uzun sürebilir! (2-3 dk)
```

```python
cart_cv_model.best_params_ # optimum degerler
```

```python
# final modelinin olusturulması ve hata degerinin bulunması,
cart_tuned = DecisionTreeRegressor(max_leaf_nodes = 4, min_samples_split = 79)
cart_tuned = cart_tuned.fit(X_train, y_train)
```

```python
ypred = cart_tuned.predict(X_test)
rmse_tuned = np.sqrt(mean_squared_error(y_test, ypred)) # hata degeri
```

```python
rmse_tuned # [önemli]
```

```python
print(f"Optimum degerler ile elde edilen modelin hata degeri: {rmse_tuned}")
```

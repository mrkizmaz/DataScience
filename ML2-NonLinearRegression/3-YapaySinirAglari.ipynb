### Verisetinin İncelenmesi ve Manipüle Edilmesi

* Kullanılan Kütüphaneler

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

* Verisetinin İncelenmesi

Verisetini [buradan](https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv) indirebilirsiniz.

```python
hit = pd.read_csv("https://mrkizmaz-s3data.s3.eu-west-1.amazonaws.com/DataSets/Hitters.csv")
df = hit.copy()
df.head()
```

```python
df.info() # veriseti hakkında bilgi verir
```

```python
df.isnull().values.any() # verisetinde bos degerler var mı?
```

```python
df = df.dropna(axis = 0) # verisetindeki bos degerlerin oldugu satırları siler
```

```python
df.info()
```

```python
df.nunique() # sütun bazında eşsiz degerlerin sayısı
```

```python
df.describe().T # verisetinin istatistiksel dagılımı
```

```python
df.corr() # aralarındaki korelasyon degerleri
```

```python
# verisetindeki kategorik degiskenler sürekli degiskenlere cevrilmeli,
dummy = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
dummy.head()
```

```python
# kategorik degiskenler ana verisetinden silinmeli,
df1 = df.drop(['League', 'Division', 'NewLeague'], axis = 1)
df1.head()
```

```python
# dummy ve df1 birlestirerek güncel veriseti olusturulur,
df = pd.concat([df1, dummy[['League_N', 'Division_W', 'NewLeague_N']]], axis = 1)
df.head()
```

### Model Kurulumu

```python
X = df.drop(['Salary'], axis =  1) # Salary dısındaki bagımsız degiskenler
y = df['Salary'] # bagımlı degisken

# %80 egitim, %20 test seti,
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.20,
                                                    random_state = 42)
# dönüstürme islemleri
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# nöron sayısı belirtilerek modelin olusturulması ve fit edilmesi,
mlp_model = MLPRegressor(hidden_layer_sizes = (100, 20)).fit(X_train_scaled, y_train)
mlp_model
```

```python
# modelin parametreleri,
```

```python
mlp_model.alpha
```

```python
mlp_model.n_layers_
```

```python
mlp_model.hidden_layer_sizes
```

```python
mlp_model.activation
```

### Tahminleme

```python
y_pred = mlp_model.predict(X_test_scaled)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred)) # hata degeri
```

```python
rmse # [önemli]
```

```python
rscore = mlp_model.score(X_train_scaled, y_train) # modelin basarı degeri
```

```python
rscore # [önemli]
```

```python
print(f"Nöron sayısı belirtilerek elde edilen modelin hata degeri: {rmse} ve Basarı yüzdesi: {rscore * 100}")
```

### Model Tuning / Model Doğrulama

```python
# optimum degerlerinin bulunması,
mlp_params = {'alpha': [0.1, 0.01, 0.02, 0.005],
             'hidden_layer_sizes': [(20, 20), (100, 50, 150), (300, 200, 150)],
             'activation': ['relu', 'logistic']}
```

```python
# modelin olusturulması ve fit edilmesi
mlp_cv_model = GridSearchCV(mlp_model, mlp_params, cv = 10)
mlp_cv_model = mlp_cv_model.fit(X_train_scaled, y_train) # bu islem 4-5 dk sürecektir
```

```python
mlp_cv_model.best_params_
```

```python
# final modelinin olusturulması,
mlp_tuned = MLPRegressor(alpha = 0.1, hidden_layer_sizes = (100, 50, 150), activation = 'relu')
mlp_tuned = mlp_tuned.fit(X_train_scaled, y_train)
```

```python
# hata degeri,
y_pred = mlp_tuned.predict(X_test_scaled)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
rmse # [önemli]
```

```python
rscore = mlp_tuned.score(X_train_scaled, y_train) # modelin anlamlılık degeri
```

```python
rscore # [önemli]
```

```python
print(f"Optimum degerler ile elde edilen hata degeri: {rmse} ve Basarı yüzdesi: {rscore * 100}")
```
